```python
# ==========================================================
# Importando bibliotecas
# ==========================================================
import os
import pandas as pd
from sqlalchemy import create_engine
from dotenv import load_dotenv, find_dotenv

# Encontra e carrega as variáveis de ambiente a partir do arquivo .env
load_dotenv(find_dotenv())

# ==========================================================
# Credenciais do Render (carregadas de forma segura)
# ==========================================================
# Use os nomes exatos do seu arquivo .env
host = os.getenv("HOST")
port = int(os.getenv("PORT"))
database = os.getenv("DATABASE")
user = os.getenv("USER")
password = os.getenv("PASSWORD")

# ==========================================================
# Criando a conexão
# ==========================================================
engine = create_engine(
    f"postgresql://{user}:{password}@{host}:{port}/{database}"
)

# ==========================================================
# Consultando os dados da tabela 'paises'
# ==========================================================
query = "SELECT * FROM fato_pais;"
df_lido = pd.read_sql(query, engine)

print("✅ Consulta realizada com sucesso!")
print(df_lido)
```
